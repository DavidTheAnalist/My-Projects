Data Cleaning in Pandas

```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\NKC TUNGA\Desktop\Python Projects\Pandas\Customer Call List.xlsx")
df
```

```python
df = df.drop_duplicates()
df
```

```python
df = df.drop(columns = 'Not_Useful_Column')
df
```

```python
df['Last_Name'] = df['Last_Name'].str.strip('123._/')
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex = True)
df
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
df
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('nan', '', regex = True)

df['Phone_Number'] = df['Phone_Number'].str.replace('Na--', '', regex = True)

#df['Phone_Number'] = df['Phone_Number'].str.strip('-')

df
```

```python
#df[['Street_Address', 'State', 'Zip_Code']] = df['Address'].str.split(',', expand=True)
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y', regex = True)

df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N', regex = True)

df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y', regex = True)

df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N', regex = True)

#df['Paying Customer'] = df['Paying Customer'].str.replace('N/a', 'N', regex = True)

df
```

```python
df = df.replace('N/a', '')
df = df.fillna('')
df
```

```python
for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] =='Y':
        df.drop(x, inplace=True)

df
```

```python
for x in df.index:
    if df.loc[x, 'Phone_Number'] =='':
        df.drop(x, inplace=True)

df

#Another way to drop null values
#df = df.dropna((subset = 'Phone_Number'), inplace=True)
```

```python
df = df.reset_index(drop=True)
df
```

```python
df = df.drop(columns = 'Address')
df
```
